```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester("HPRMToUrA4F7voa8") as m:
    structure = m.get_structure_by_material_id("mp-961649")
    
structure.to(filename="ZrFeTe.cif")
```

```python
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.core.operations import SymmOp
from IPython.display import Image

#view = quick_view(structure)
#display(view)
```

```python
# Opérations de symétrie
analyzer = SpacegroupAnalyzer(structure).get_symmetry_operations()
S2 = analyzer[1]
S3 = analyzer[2]
S6 = analyzer[5]

# Atomes
Zr = [0.5000, 0.5000, 0.5000] #1
Fe = [0.2500, 0.2500, 0.2500] #2
Te = [0.5000, 0.5000, 0.0000] #5

# Application des opérations de symétrie
Zr2 = S2.operate(Zr)
Fe2 = S3.operate(Fe)
Te2 = S6.operate(Te)
```

```python
print("Opération de symétrie sur Zr: axe 4-bar|centre d'inversion en 0 0 0 (y,z,-x-y-z)\n")
print(S2,"\n")
print("Atome de départ : ",Zr,"\n")
print("Atome après symétrie : ",Zr2,"\n")

view = quick_view(structure)
display(view)
view.script("draw symop 2 {atomno=1}")
Image(filename='S2.png')
```

```python
print("Opération de symétrie sur Fe: axe C2 (-x-y-z,z,y)\n")
print(S3,"\n")
print("Atome de départ : ",Fe,"\n")
print("Atome après symétrie : ",Fe2,"\n")

view = quick_view(structure)
display(view)
view.script("draw symop 3 {atomno=2}")
Image(filename='S3.png')
```

```python
print("Opération de symétrie sur Te: plan miroir (x,-x-y-z,z)\n")
print(S6,"\n")
print("Atome de départ : ",Te,"\n")
print("Atome après symétrie : ",Te2,"\n")

view = quick_view(structure)
display(view)
view.script("draw symop 6 {atomno=9}")
Image(filename='S6.png')
```
